```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib as plt
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['Omicron_ - Copy.csv']))
df
```

```python
data = df.groupby("Country", as_index=False).last()

fig = px.choropleth(
    data,
    color="Percentage of Omicron",
    labels={"Percentage of Omicron"},
    locations="Country",
    locationmode="country names",
    custom_data=["Date"],
    color_continuous_scale=["#F0DC82", "gold", "red"],
    title="Percentage of Omicron Cases by Country",
)
fig.show()
```

```python
peak_cases = df.groupby('Country')[['Percentage of Omicron']].max().sort_values(['Percentage of Omicron'], ascending = [True]).reset_index()
peak_cases
```

```python

fig_peak = px.bar(peak_cases, x='Country', y='Percentage of Omicron')
fig_peak.show()
```

```python
df_germany = df.loc[df['Country']=='Germany']
df_germany
```

```python
Germany = df_germany.groupby(['Date'])[['Percentage of Omicron']].sum().sort_values(['Percentage of Omicron'], ascending = [True])
Germany
```

```python
fig1 = px.line(df_germany, x='Date', y='Percentage of Omicron', title ='Trend of Omicron cases in Germany')
fig1.show()
```
